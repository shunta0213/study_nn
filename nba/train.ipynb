```python
player_data = pd.read_csv("ff_player.csv")

# Drop columns: player, team, gb(game behind)
features = player_data.drop(columns=["Player", "Team", "GB"])

# categorical variables encoding
# there are some methods: One-hot-encoding(used), Label-encoding, Count-encoding, Target-encoding
features = pd.get_dummies(features, columns=["Pos"])

features = features.fillna(0)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

target = player_data["Rank"]

features.head(10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    scaled_features, target, test_size=0.2
)

X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32)

batch_size = 32
train_dataset = torch.utils.data.TensorDataset(X_train_tensor, y_train_tensor)
train_loader = torch.utils.data.DataLoader(
    dataset=train_dataset, batch_size=batch_size, shuffle=True
)
```

```python
import torch.nn as nn


class BasketballMVPModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(BasketballMVPModel, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x
```
